```
df = pd.read_excel('data/16k_apperals_data.xlsx', index_col=False)
```

```
df.head()
```

```
categories = df.product_type_name.unique()
brands = df.brand.unique()
colors = df.color.unique()
```

```
category2id = {o:i for i,o in enumerate(categories)}
brand2id = {o:i for i,o in enumerate(brands)}
color2id = {o:i for i,o in enumerate(colors)}


id2category = {v:k for k,v in category2id.items()}
id2brand = {v:k for k,v in brand2id.items()}
id2color = {v:k for k,v in color2id.items()}
```

```
id2category
```

```
df['category_id'] = df.product_type_name.apply(lambda x: category2id[x])
df['brand_id'] = df.brand.apply(lambda x: brand2id[x])
df['color_id'] = df.color.apply(lambda x: color2id[x])
```

```
df.head()
```

```
df.to_excel('16k_apperals_database.xlsx')
```
